```python

plt.figure()
for key in center_lines.keys():
    plot_over_line = pa.pa.PlotOverLine(filter)
    plot_over_line.Source.Point1 = center_lines[key][0]
    plot_over_line.Source.Point2 = center_lines[key][1]
    data = pa.save_data_to_dict(plot_over_line,cell=False)
    plt.plot(data["arc_length"],data["P"],label=key,lw=2)
    
plt.legend(loc='upper right',fancybox=True, shadow=True)
plt.grid()
plt.xlabel("Length [m]")
plt.ylabel("Pressure [Pa]")
plt.show()
```
